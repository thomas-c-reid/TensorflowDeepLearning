```python
import pandas as pd

data = {
    'Name': ['Anne','Bob','Sarah','Jane','Marc'],
    'Age': [20,45,19,None,33],
    'Gender': ['f','m','f','f','m'],
    'Job': ['programmer','writer','cook','programmer','teacher'],
    'Income': [50000, 60000, 30000, 400000, None]
}

df = pd.DataFrame(data)
df
```

```python
# Proprocessing steps

# . Drop Name Features
# . Impute ages
# . Turn gender into binary
# . one hot encode jobs
# . scale income column
```

```python
from  sklearn.base import BaseEstimator, TransformerMixin
from sklearn.impute import SimpleImputer

class NameDropper(BaseEstimator, TransformerMixin):

  def fit(self, X, y=None):
    return self

  def transform(self, X):
    return X.drop(['Name'], axis=1)

class AgeImputer(BaseEstimator, TransformerMixin):

  def fit(self, X, Y=None):
    return self

  def transform(self, X):
    imputer = SimpleImputer(strategy='mean')
    X['Age'] = imputer.fit_transform(X[['Age']])
    return X
```

```python
data2 = {
    'Name': ['Kasey','Mike','laura','Jane','issac'],
    'Age': [23,14,26,None,33],
    'Gender': ['f','m','f','f','m'],
    'Job': ['programmer','writer','cook','programmer','teacher'],
    'Income': [50000, None, 30000, 400000, None]
}

df2 = pd.DataFrame(data2)
df2

dropper = NameDropper()
imp = AgeImputer()

imp.fit_transform(dropper.fit_transform(df2))
```

```python
data2 = {
    'Name': ['Kasey','Mike','laura','Jane','issac'],
    'Age': [23,14,26,None,33],
    'Gender': ['f','m','f','f','m'],
    'Job': ['programmer','writer','cook','programmer','teacher'],
    'Income': [50000, None, 30000, 400000, None]
}

df2 = pd.DataFrame(data2)
df2

from sklearn.pipeline import Pipeline

pipe = Pipeline([
    ('Dropper', NameDropper()),
    ('imputer', AgeImputer())
])

pipe.fit_transform(df2)
```
